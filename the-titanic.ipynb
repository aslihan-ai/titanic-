## Part 2: Titanic Dataset Analysis

```python
df = pd.read_csv("/kaggle/input/datasets/nesihamohammed/titanic-dataset/dataset-2.csv")
df.head()
```

### Data Exploration

```python
df.head()
df.info()
df.describe()
```

### Data Cleaning
Handling missing values and fixing column names

```python
# Fix column names (important for this dataset)
df.columns = df.columns.str.strip().str.title()

# Fill Age with median
if "Age" in df.columns:
    df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill Embarked with mode (if exists)
if "Embarked" in df.columns:
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin if exists
if "Cabin" in df.columns:
    df = df.drop("Cabin", axis=1)

# Remove duplicates
df = df.drop_duplicates()

df.info()
```

### Data Analysis using groupby()

```python
# Survival rate by gender
df.groupby("Sex")["Survived"].mean()

# Survival rate by class
df.groupby("Pclass")["Survived"].mean()

# Average age per class
df.groupby("Pclass")["Age"].mean()
```

```python
df["AgeGroup"] = pd.cut(df["Age"],
                       bins=[0, 12, 18, 60, 100],
                       labels=["Child", "Teen", "Adult", "Senior"])

df.groupby("AgeGroup", observed=True)["Survived"].mean()
```

### Data Filtering

```python
# Female survivors
df[(df["Sex"] == "Female") & (df["Survived"] == 1)]

# Children survivors
df[(df["Age"] < 12) & (df["Survived"] == 1)]

# First class survivors
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

## Insights

- Females were more likely to survive than males.

- Passenger class significantly affected survival. First-class passengers had the highest survival rate.

- Children had relatively higher survival rates, indicating they were prioritized.

- The highest survival probability was observed among female passengers in first class, especially younger individuals.
